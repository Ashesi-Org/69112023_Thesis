```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## Balanced vs Unbalanced Data

### Data Points

```python
#X-axis applies to all
x = [1,2,3]
x = np.linspace(1,3,3)

#KNN
# ub_y1 = [0.1,0.72,0.21]
ub_y1 = (0.1,0.72,0.21)
# b_y1 = [0.46,0.35,0.42]
b_y1 = (0.46,0.35,0.42)

#RF
# ub_y2 = [0.56,0.77,0.66]
ub_y2 = (0.56,0.77,0.66)
# b_y2 = [0.81,0.58,0.71]
b_y2 = (0.81,0.58,0.71)

#ERF
ub_y3 = [0.5,0.71,0.62]
b_y3 = [0.75,0.53,0.66]

#MLP
ub_y4 = [0.41,0.75,0.59]
b_y4 = [0.77,0.54,0.62]
```

### Plot

```python
# Scatter plot unbalanced data values
for i in range(3):
    plt.scatter(x[i], ub_y1[i], marker='o', color='r')
    plt.scatter(x[i], ub_y2[i], marker='o', color='g')
    plt.scatter(x[i], ub_y3[i], marker='o', color='b')
    plt.scatter(x[i], ub_y4[i], marker='o', color='y')

    
# Scatter plot balanced data values
for i in range(3):
    plt.scatter(x[i], b_y1[i], marker='s', color='r')
    plt.scatter(x[i], b_y2[i], marker='s', color='g')
    plt.scatter(x[i], b_y3[i], marker='s', color='b')
    plt.scatter(x[i], b_y4[i], marker='s', color='y')

# Add labels and title
plt.xlabel('Damage Grade')
plt.ylabel('Accuracy')
plt.title('Balanced vs Unbalanced Accuracy')

# Display the plot
plt.show()
```

```python
# Create a line graph for y1
plt.plot(x, ub_y1, 'r-', label='ub knn')

# Create a line graph for y2
plt.plot(x, b_y1, 'b-', label='b knn')

# Add labels and title to the plot
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Graph')

# Add a legend to the plot
plt.legend()

# Display the plot
plt.show()
```
